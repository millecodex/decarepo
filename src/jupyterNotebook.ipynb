### Interactive Environment
Jupyter Lab is great for development and data science because output appears immediately below the code block:

```python
# lets output some random numbers!
import numpy as np
rng = np.random.default_rng(seed=42)
array = rng.random((100, 1))
array[:20]
```

### Plotting
We can also display plots in the notebook without calling external windows/plotters

```python
import matplotlib.pyplot as plt
 
# data to be plotted
x = np.arange(1, 101)
y = array
 
# plotting
plt.title("Random Sampling")
plt.xlabel("sample")
plt.ylabel("value")
plt.plot(x, y, color ="red")
plt.show()
```
